```python
top_complaints = df['Complaint Type'].value_counts().nlargest(10)

# Ploting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_complaints.index, top_complaints.values, color='Pink')
plt.title('Top 10 Complaint Types in NYC')
plt.xlabel('Complaint Type')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=45, ha="right")  
plt.show()
```

```python
df['Month'] = df['Created Date'].dt.month
```

```python
monthly_complaints = df['Month'].value_counts().sort_index()

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(monthly_complaints.index, monthly_complaints.values, marker='o', color='skyblue')
plt.title('Monthly Complaint Trend in NYC')
plt.xlabel('Month')
plt.ylabel('Number of Complaints')
plt.xticks(range(1, 13))  # Shows all months from 1 to 12
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
avg_closing_time = df.groupby('Complaint Type')['Request_Closing_Time (in hours)'].mean().nlargest(10)
```

```python
# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(avg_closing_time.index, avg_closing_time.values, color='Orange')
plt.title('Average Request Closing Time by Complaint Type')
plt.xlabel('Complaint Type')
plt.ylabel('Average Closing Time (Hours)')
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Counting the number of complaints per Borough
borough_counts = df['Borough'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(borough_counts, labels=borough_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart of Complaints by Borough')
plt.axis('equal')  # Makes the pie chart circular
plt.show()
```

```python
# Group by Complaint Type and Location Type, then calculate average closing time
grouped = df.groupby(['Complaint Type', 'Borough'])['Request_Closing_Time (in hours)'].mean().reset_index()

# Pivot to get Complaint Types as rows and Location Types as columns
pivot_table = grouped.pivot(index='Complaint Type', columns='Borough', values='Request_Closing_Time (in hours)')

# Get top 8 complaint types by overall average closing time
top_complaints = pivot_table.mean(axis=1).sort_values(ascending=False).head(8)
pivot_table = pivot_table.loc[top_complaints.index]

# Plot grouped bar chart with thicker bars
ax = pivot_table.plot(kind='bar', figsize=(12, 6))

# Chart styling
plt.title('Average Closing Time by Complaint and Borough', fontsize=14, fontweight='bold')
plt.ylabel('Avg Closing Time (hours)', fontsize=12)
plt.xlabel('Complaint Type', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Location Type', fontsize=9, title_fontsize=10)
plt.tight_layout()

plt.show()
```

```python
from scipy import stats
```

```python
# Pick a few common complaint types with enough data
top_complaints = df['Complaint Type'].value_counts().head(10).index

# Extract closing times for each type
groups = [df[df['Complaint Type'] == ct]['Request_Closing_Time (in hours)'].dropna() for ct in top_complaints]

# Perform one-way ANOVA
f_stat, p_value = stats.f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject H0: Average response times are significantly different across complaint types.")
else:
    print("Fail to reject H0: No significant difference in average response times across complaint types.")
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Filter to top 10 complaint types and top 5 boroughs for simplicity
top_complaints = df['Complaint Type'].value_counts().head(10).index
top_boroughs = df['Borough'].value_counts().head(5).index
filtered_df = df[df['Complaint Type'].isin(top_complaints) & df['Borough'].isin(top_boroughs)]

# Create the contingency table
contingency_table = pd.crosstab(filtered_df['Complaint Type'], filtered_df['Borough'])

# Apply chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Contingency Table:\n", contingency_table)

print("Chi-Square Statistic:", chi2)
print("P-value:", p)

# Interpretation
if p < 0.05:
    print("There is a statistically significant relationship between complaint type and borough.")
else:
    print("There is no statistically significant relationship between complaint type and borough.")

```
